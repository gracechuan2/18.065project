```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("dataset.csv")
df = df.drop(['Tier', 'country', 'country_abb'], axis=1)
df.head()
```

```python
# # some of the rows have 'NaN' as labels, so gotta drop those
drop_row_indices = []
for i, row in df.iterrows():
    for j, col in row.iteritems():
        try:
            row[j] = float(col)
        except ValueError:
            print("ValueError")
            drop_row_indices.append(i)
            continue
        if math.isnan(row[j]):
            drop_row_indices.append(i)

```

```python
df = df.drop(drop_row_indices, axis=0)
```

```python
features = ['year', 'VAA', 'PSNV', 'GE', 'RQ', 'RL', 'CC']

X = df.loc[:, features]
y = df.GDPgrowth
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15)

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
predicted_y = regression_model.predict(X_test)
mean_squared_error(y_test, predicted_y)
```
